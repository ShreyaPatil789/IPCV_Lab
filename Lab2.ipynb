```python
img = cv2.imread("img1.jpg", 0)
```

```python
negative_img = 255 - img
```

```python
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Image Negative")
plt.imshow(negative_img, cmap='gray')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np


img = cv2.imread("img1.jpg", cv2.IMREAD_GRAYSCALE)



L = 256

# Negative image formula: s = L - 1 - r
negative = (L - 1) - img

# Print grid values
print("Original Image Grid Values (r):")
print(img)

print("\nNegative Image Grid Values (s = L - 1 - r):")
print(negative)

# Save result
cv2.imwrite("negative_image.jpg", negative)
```

```python
region = img[:8, :8]

print("Original Pixel Grid:\n", region, "\n")


for k in range(8):
    bit_plane = (region >> k) & 1
    print(f"Bit Plane {k} Grid Values:\n{bit_plane}\n")
    plt.subplot(2, 4, k+1)
    plt.imshow(bit_plane * 255, cmap='gray')
    plt.title(f"Bit Plane {k}")
    plt.axis('off')
```
